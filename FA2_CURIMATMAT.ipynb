```r

u <- c(3,-2,5)
v <- c(-4,-1,6)


a <- sum(u * v)

b <- sum((u +3) * (v -4))

c <- sum(v^2)

d <- sum(u * (v^2))

e <- sum(u*v^2)

f <- sum(v^2 - 2*v^2 + 2)

g <- sum(u / v)


cat("(a):", a, "\n")
cat("(b):", b, "\n")
cat("(c):", c, "\n")
cat("(d):", d, "\n")
cat("(e):", e, "\n")
cat("(f):", f, "\n")
cat("(g):", round(g, 4), "\n")
```

```r
set_a <- c(3, 5, 8, 3, 7, 2)
set_b <- c(28.5, 73.6, 47.2, 31.5, 64.8)

gm_a <- exp(mean(log(set_a)))
gm_b <- exp(mean(log(set_b)))

cat("Geometric Mean (a):", round(gm_a, 2), "\n")
cat("Geometric Mean (b):", round(gm_b, 2), "\n")
```
