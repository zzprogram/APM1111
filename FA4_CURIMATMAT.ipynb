```r
data <- read.csv("Data-1.csv")

# function to calculate
get_moms <- function(x, point) {
  sapply(1:4, function(r) mean((x - point)^r))
}

# 1. raw Moments
raw_moms <- sapply(data, get_moms, point = 0)

# 2. moments about Mean
mean_moms <- sapply(data, function(x) get_moms(x, mean(x)))

# 3. moments about the number 75 (female height measurements)
num_75 <- get_moms(data$Normal, 75)

# output
rownames(raw_moms) <- c("1st", "2nd", "3rd", "4th")
rownames(mean_moms) <- c("1st", "2nd", "3rd", "4th")

# 4. verify the relations between the moments

# define the values (prob 1)
m1_prime <- 65.12
m2_prime <- 4248.92
m3_prime <- 277770.92
m4_prime <- 18194173.64

# define the values (prob 3)
ma1 <- -9.88
ma2 <- 105.92
ma3 <- -1211.08
ma4 <- 14572.64

# calculate prob 2 using prob 1
# prob 1 (m') to prob 2 (M)
Ma_1 <- 0
Ma_2 <- m2_prime - (m1_prime^2)
Ma_3 <- m3_prime - 3*m1_prime*m2_prime + 2*(m1_prime^3)
Ma_4 <- m4_prime - 4*m1_prime*m3_prime + 6*(m1_prime^2)*m2_prime - 3*(m1_prime^4)

# calculate prob 2 using prob 3
Mb_1 <- 0
Mb_2 <- ma2 - (ma1^2)
Mb_3 <- ma3 - 3*ma1*ma2 + 2*(ma1^3)
Mb_4 <- ma4 - 4*ma1*ma3 + 6*(ma1^2)*ma2 - 3*(ma1^4)

# comparison table
comparison <- data.frame(
  Derived_Prob1 = c(Ma_1, Ma_2, Ma_3, Ma_4),
  Derived_Prob3 = c(Mb_1, Mb_2, Mb_3, Mb_4)
)


print("Raw Moments:")
print(raw_moms)
cat("\n\n")
print("Moments about Mean:")
print(mean_moms)
cat("\n\n")
print("Moments about number 75 (Female Heights Measurements):")
print(num_75)
cat("\n\n")
print("Verify the Relations:")
print(comparison)
```
