```r
# dataset
males <- c(12, 4, 11, 13, 11, 7, 9, 10, 10, 7, 7, 12, 6, 9, 15, 10, 11, 12, 7,
           8, 8, 9, 11, 10, 9, 10, 9, 9, 7, 9, 11, 7, 10, 10, 11, 9, 12, 12, 8,
           13, 9, 10, 8, 11, 10, 13, 13, 9, 10, 13)

females <- c(11, 9, 7, 10, 9, 10, 10, 7, 9, 10, 11, 8, 9, 6, 11, 10, 7, 9, 12,
             14, 11, 12, 12, 8, 12, 12, 9, 10, 11, 7, 12, 7, 9, 8, 11, 10, 8,
             13, 8, 10, 9, 9, 9, 11, 9, 9, 8, 9, 12, 11)

all_data <- c(males, females)

# 1 - entire dataset

# calculate statistics
n <- length(all_data)
mean_val <- mean(all_data)
sd_val <- sd(all_data)
error <- qt(0.975, df=n-1) * sd_val/sqrt(n)

# table
q1_table <- data.frame(
  Statistic = c("Sample Size (n)", "Mean", "Median", "Standard Deviation",
  "Variance", "Minimum", "Maximum", "95% CI Lower", "95% CI Upper"),
  Value = c(n, mean_val, median(all_data), sd_val, var(all_data),
  min(all_data), max(all_data), CI_Lower=mean_val - error,
  CI_Upper= mean_val + error)
)


# 2 - by gender

#get stats
get_group_stats <- function(x) {
  n <- length(x)
  error <- qt(0.975, df=n-1) * sd(x)/sqrt(n)
  c(Mean=mean(x), Median=median(x), SD=sd(x), Var=var(x), Min=min(x),
  Max=max(x), CI_Lower=mean(x)-error, CI_Upper=mean(x)+error)
}

# table
comparison_table <- data.frame(
  Statistic = c("Mean", "Median", "SD", "Variance", "Min", "Max",
  "95% CI Lower", "95% CI Upper"),
  Males = get_group_stats(males),
  Females = get_group_stats(females)
)

#ouput

cat("ENTIRE DATASET\n")
print(q1_table, row.names = FALSE)
cat("\nQuartiles:\n")
print(quantile(all_data, probs = c(0.25, 0.5, 0.75)))
cat("\nDeciles:\n")
print(quantile(all_data, probs = seq(0.1, 0.9, by=0.1)))

cat("\n\nGROUPED BY GENDER\n")
print(comparison_table, row.names = FALSE)
cat("\nDeciles - Males:\n")
print(quantile(males, probs = seq(0.1, 0.9, by=0.1)))
cat("\nDeciles - Females:\n")
print(quantile(females, probs = seq(0.1, 0.9, by=0.1)))
```

**3. Statistical Report**

1. Overview of the Entire Dataset

The combined sample consists of 100 college students (50 males and 50 females). The primary goal to know how many hours per week these students spend talking on their cell phones.


*   The overall average (mean) usage is 9.79 hours per week, with a median of 10.0. Since these two values are nearly identical, it indicates that the distribution is very balanced and not heavily skewed by extreme outliers.
*   The standard deviation is 1.97, and the variance is 3.86. This tells us that most students (roughly 68%) fall within approximately 2 hours of the mean (between 7.8 and 11.8 hours).
* Usage varies from a minimum of 4 hours to a maximum of 15 hours.

2. Comparison Across Gender Categories

When splitting the data by gender, we see that the habits of male and female students are remarkably similar, though subtle has differences:

* Consistency: Females show a smaller standard deviation (1.78) compared to males (2.15). Suggesting that female phone usage is slightly more consistent or clustered around the average, than the male.

3. Interpretation of 95% Confidence Intervals

The 95% Confidence Interval (CI) estimates where the true population mean likely sits for the entire university.

* Overall CI: [9.40, 10.18] hours.
* Male CI: [9.27, 10.49] hours.
* Female CI: [9.20, 10.20] hours.

There is a significant overlap between the male and female confidence intervals. In statistics, when intervals overlap this much, it suggests that there is no significant evidence to claim that males and females talk for different amounts of time. The small difference in our sample means is likely due to random sampling variation rather than a fundamental difference in gender behavior.

4. Patterns and Observations

* Both groups show very low skewness, meaning the data follows a "bell-shaped" pattern fairly well.
* Looking at the deciles, the 20th percentile (8.0) and the 80th percentile (11.2) are quite close. This shows that the vast majority of students—regardless of gender—are heavy users of cell phones, typically talking between 8 and 11 hours per week.
* The male group contains both the absolute minimum (4) and the absolute maximum (15), indicating that the most "extreme" phone users (both low and high) in this specific study were male.

```r
# dataset
x <- c(2, 3, 7, 8, 10, 10, 12, 15, 17, 20)
n <- length(x)

# 1. raw moments (m')
m1_prime <- mean(x)
m2_prime <- mean(x^2)
m3_prime <- mean(x^3)
m4_prime <- mean(x^4)

# 2. moments about the mean (m)
m1 <- mean((x - m1_prime)^1)
m2 <- mean((x - m1_prime)^2)
m3 <- mean((x - m1_prime)^3)
m4 <- mean((x - m1_prime)^4)

# 3. verify
calculated_m4 <- m4_prime - 4*m1_prime*m3_prime + 6*(m1_prime^2)*m2_prime -
  3*(m1_prime^4)

# output
cat("Raw Moments:\n")
cat("m1':", m1_prime, "\nm2':", m2_prime, "\nm3':", m3_prime, "\nm4':",
    m4_prime, "\n\n")

cat("Moments about the Mean:\n")
cat("m1:", round(m1, 10), "(essentially 0)\n")
cat("m2:", m2, "\nm3:", m3, "\nm4:", m4, "\n\n")

cat("Verify:\n")
cat("Directly calculated m4: ", m4, "\n")
cat("Formula derived m4:   ", calculated_m4, "\n")
cat("Are they equal?       ", all.equal(m4, calculated_m4), "\n")
```
