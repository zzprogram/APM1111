```r
# define the population
population <- c(9, 12, 15)
n <- 2

# all possible samples of size n=2 with replacement (a data frame with all combinations)
samples <- expand.grid(Value1 = population, Value2 = population)

# calculate the mean for each sample (row)
samples$Mean <- rowMeans(samples)

# sampling distribution table
sampling_dist <- as.data.frame(table(samples$Mean))
colnames(sampling_dist) <- c("x_bar", "Frequency")
sampling_dist$Probability <- sampling_dist$Frequency / sum(sampling_dist$Frequency)

# verification of means (part c)
mu_x_bar <- sum(as.numeric(as.character(sampling_dist$x_bar)) * sampling_dist$Probability)

# verification of variance (part d)
x_bar_num <- as.numeric(as.character(sampling_dist$x_bar))
sigma2_x_bar <- sum(x_bar_num^2 * sampling_dist$Probability) - (mu_x_bar^2)

# results
cat("All 9 Possible Samples","\n")
print(samples)
cat("\nMean of Sampling Distribution :", mu_x_bar)
cat("\nVariance of Sampling Distribution:", round(sigma2_x_bar, 2))
```
