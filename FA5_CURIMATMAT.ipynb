```r
# (a)
dice <- expand.grid(d1 = 1:6, d2 = 1:6, d3 = 1:6)
sums <- rowSums(dice)
dist_table <- table(sums) / length(sums)
cat("Probability Distribution of x:")
print(dist_table)

cat("\n\n")

# (b)
target <- c("7", "8", "9", "10")
prob_2b <- sum(dist_table[names(dist_table) %in% target])
cat("Probability of getting 7, 8, 9, or 10:", "\n", prob_2b)
```

```r
x <- c(-10, -20, 30)
p <- c(1/5, 3/10, 1/2)

# (a) E(X)
ex <- sum(x * p)

# (b) E(X^2)
ex2 <- sum(x^2 * p)

# (c) Variance E[(X-mean)^2]
variance <- ex2 - (ex^2)

# (d) E(X^3)
ex3 <- sum(x^3 * p)

cat("(a):", ex, "\n")
cat("(b):", ex2, "\n")
cat("(d):", ex3, "\n")
```

```r
# (a)
# (6-1)!
n_total <- 6
ans_a <- factorial(n_total - 1)

# (b)
# if 2 women are 1 unit, we arrange 5 units: (5-1)! * 2!
together <- factorial(5 - 1) * factorial(2)
ans_b <- ans_a - together

# (c)
# fix one man: (3-1)! ways to seat men. 3 gaps for 3 women: 3! ways.
ans_c <- factorial(3 - 1) * factorial(3)

cat("(a):", ans_a, "\n(b):", ans_b, "\n(c):", ans_c)
```

```r
# (a)
# select 2 men(6), 4 women(8), 3 boys(4), 3 girls(5)
ans_a <- choose(6, 2) * choose(8, 4) * choose(4, 3) * choose(5, 3)

# (b)
# need 1 more man from 5, and 3 more women from 7
ans_b <- choose(5, 1) * choose(7, 3) * choose(4, 3) * choose(5, 3)

cat("(a):", ans_a, "\n(b):", ans_b)
```
